```python
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print("Popped:", stack.pop())
print("Stack:", stack)
```

```python
queue = []

queue.append(10)
queue.append(20)
queue.append(30)

queue[1] = 25

print("Removed:", queue.pop(0))
print("Queue:", queue)
```

```python
arr = []

arr.append(1)
arr.append(2)
print("Stack Pop:", arr.pop())

arr.append(3)
arr.append(4)
print("Queue Dequeue:", arr.pop(0))

print("Final List:", arr)
```

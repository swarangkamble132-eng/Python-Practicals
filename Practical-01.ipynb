```python
R1 = float(input("Enter first resistance: "))
R2 = float(input("Enter second resistance: "))
Rs = R1 + R2
print("Equivalent Resistance in Series =", Rs, "ohms")
```

```python
R1 = float(input("Enter first resistance: "))
R2 = float(input("Enter second resistance: "))
Rp = (R1 * R2) / (R1 + R2)
print("Equivalent Resistance in Parallel =", Rp, "ohms")
```

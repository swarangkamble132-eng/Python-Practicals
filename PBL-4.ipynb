1. Linear Search
   
Develop a Python program to implementLinear Search to find a given element from a list of academic records.



```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

records = [101, 104, 108, 110, 115]
key = 110

result = linear_search(records, key)
print("Element found at index:", result)
```

2. Binary Search
Develop a Python program to implementBinary Search on a sorted list of student roll numbers.

```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1

roll_numbers = [1, 3, 5, 7, 9, 11]
key = 7

print("Element found at index:", binary_search(roll_numbers, key))
```

3. Bubble Sort
  
  Write a Python program to implement Bubble Sort to arrange marks of students in ascending or descending order.




```python
marks = [55, 72, 68, 90, 60]

for i in range(len(marks)):
    for j in range(len(marks)-1):
        if marks[j] > marks[j+1]:
            marks[j], marks[j+1] = marks[j+1], marks[j]

print("Sorted Marks:", marks)
```

4. Selection Sort
   
Write a Python program to implementSelection Sort to sort laboratory equipment IDs.

```python
equipment = [23, 11, 45, 7, 30]

for i in range(len(equipment)):
    min_index = i
    for j in range(i+1, len(equipment)):
        if equipment[j] < equipment[min_index]:
            min_index = j
    equipment[i], equipment[min_index] = equipment[min_index], equipment[i]

print("Sorted Equipment IDs:", equipment)
```

5. Insertion Sort

Write a Python program to implementInsertion Sort to organize student data efficiently.


```python
students = [45, 20, 60, 10, 30]

for i in range(1, len(students)):
    key = students[i]
    j = i - 1
    while j >= 0 and key < students[j]:
        students[j + 1] = students[j]
        j -= 1
    students[j + 1] = key

print("Sorted Student Data:", students)
```

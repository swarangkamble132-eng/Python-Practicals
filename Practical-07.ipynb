```python
accounts = {}

n = int(input("Enter number of accounts: "))
for i in range(n):
    acc = int(input("Enter account number: "))
    bal = int(input("Enter balance: "))
    accounts[acc] = bal

print("Accounts:", accounts)

print("Total balance:", sum(accounts.values()))
print("Average balance:", sum(accounts.values()) / len(accounts))

print("Highest balance:", max(accounts.values()))
print("Lowest balance:", min(accounts.values()))

limit = int(input("Enter high balance limit: "))
print("High value accounts:", {k:v for k,v in accounts.items() if v >= limit})

search = int(input("Enter account number to search: "))
print("Found" if search in accounts else "Not Found")

acc = int(input("Enter account number to update: "))
accounts[acc] = int(input("Enter new balance: "))
print("After update:", accounts)

acc = int(input("Enter account number to delete: "))
accounts.pop(acc, None)
print("After deletion:", accounts)

print("Account numbers:", list(accounts.keys()))
print("Balances:", list(accounts.values()))

print("Sorted accounts:", dict(sorted(accounts.items())))
```

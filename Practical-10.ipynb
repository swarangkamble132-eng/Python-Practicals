```python
import math

Vm = 20   # peak voltage
Vdc = Vm / math.pi

print("Peak Voltage (Vm):", Vm)
print("DC Output Voltage (Vdc):", Vdc)
```

```python
import statistics

data = [12, 15, 20, 20, 25, 30]

print("Data:", data)
print("Mean:", statistics.mean(data))
print("Median:", statistics.median(data))
print("Mode:", statistics.mode(data))
print("Standard Deviation:", statistics.stdev(data))
```

```python
# User-defined calculator module (in same file)

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b

# Using the calculator functions
print("Addition:", add(10, 5))
print("Subtraction:", sub(10, 5))
print("Multiplication:", mul(10, 5))
print("Division:", div(10, 5))
```

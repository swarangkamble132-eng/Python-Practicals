2. User-Defined Functions

Develop a Python program using user-defined functions to implement Half Adder and Full Adder logic.


```python

def half_adder(a, b):
    sum = a ^ b
    carry = a & b
    print("Half Adder -> Sum:", sum, "Carry:", carry)


def full_adder(a, b, c):
    sum = a ^ b ^ c
    carry = (a & b) | (b & c) | (a & c)
    print("Full Adder -> Sum:", sum, "Carry:", carry)


half_adder(1, 1)
full_adder(1, 1, 1)

```

3. Classes and Objects

Write a Python program to define a Student class with appropriate data members and member functions.
Create objects of the class and display student details.


```python
class Student:
    def __init__(self, roll, name, marks):
        self.roll = roll
        self.name = name
        self.marks = marks

    def display(self):
        print("Roll No:", self.roll)
        print("Name:", self.name)
        print("Marks:", self.marks)
        print()

# Creating objects
s1 = Student(1, "Rahul", 85)
s2 = Student(2, "Sneha", 90)

# Display student details
s1.display()
s2.display()
```


5. User-Defined Module

Create and use a user-defined module (for example, a calculator module) to perform arithmetic operations.

```python
# User-defined calculator functions

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    return a / b

# Using the functions
print("Addition:", add(10, 5))
print("Subtraction:", sub(10, 5))
print("Multiplication:", mul(10, 5))
print("Division:", div(10, 5))
```


6. NumPy Package

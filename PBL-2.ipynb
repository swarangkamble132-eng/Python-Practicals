1. List Operations
Develop a Python program to create, access, update, and delete student records using a List data structure.

```python
students = []

n = int(input("Enter number of students: "))
for i in range(n):
    name = input("Enter student name: ")
    students.append(name)

print("Student List:", students)

students.append(input("Enter name to add: "))
print("After adding:", students)

students.remove(input("Enter name to delete: "))
print("After deleting:", students)
```

2. Tuple Operations
Write a Python program to store fixed laboratory information using a Tuple and perform access and modification operations.

```python
lab = (
    input("Enter lab name: "),
    int(input("Enter capacity: ")),
    input("Enter status: ")
)

print("Lab Details:", lab)

lab_list = list(lab)
lab_list[2] = input("Enter new status: ")
lab = tuple(lab_list)

print("Updated Lab Details:", lab)
```

3. Set Operations
Develop a Python program to manage unique laboratory equipment records using a Set and perform add and remove operations.

```python
equipment = set()

n = int(input("Enter number of equipment: "))
for i in range(n):
    equipment.add(input("Enter equipment name: "))

print("Equipment:", equipment)

equipment.add(input("Enter equipment to add: "))
print("After adding:", equipment)

equipment.remove(input("Enter equipment to remove: "))
print("After removing:", equipment)
```

4. Dictionary Operations
Write a Python program to store and manage student details (Roll No, Name, Marks) using a Dictionary, including create, access, update, delete, and traversal operations.


```python
students = {}

n = int(input("Enter number of students: "))
for i in range(n):
    roll = int(input("Enter roll no: "))
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    students[roll] = [name, marks]

print("Student Records:", students)

roll = int(input("Enter roll no to update marks: "))
students[roll][1] = int(input("Enter new marks: "))
print("After update:", students)

roll = int(input("Enter roll no to delete: "))
students.pop(roll)
print("After deletion:", students)
```

5. Student Record Management
Design a Python program using appropriate data structures to maintain and display multiple student records.

```python
records = []

n = int(input("Enter number of students: "))
for i in range(n):
    name = input("Enter name: ")
    roll = int(input("Enter roll: "))
    marks = int(input("Enter marks: "))
    records.append((name, roll, marks))

for r in records:
    print("Name:", r[0], "Roll:", r[1], "Marks:", r[2])
```

6. Laboratory Record Handling
Develop a Python program to manage laboratory resources and availability using Python data structures.

```python
labs = {}

n = int(input("Enter number of labs: "))
for i in range(n):
    name = input("Enter lab name: ")
    status = input("Enter status: ")
    labs[name] = status

for lab, status in labs.items():
    print(lab, ":", status)
```

7. Data Traversal
Write a Python program to traverse and display all stored student and laboratory records efficiently.



```python
for lab in labs:
    print("Lab:", lab, "Status:", labs[lab])

```

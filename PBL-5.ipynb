1. Stack Implementation

Write a Python program to implement a Stack using array with PUSH and POP operations for managing records.



```python
stack = []

def push(item):
    stack.append(item)
    print("Pushed:", item)

def pop():
    if len(stack) == 0:
        print("Stack is empty")
    else:
        print("Popped:", stack.pop())

push(10)
push(20)
push(30)
pop()
pop()
```

2. Queue Implementation

Develop a Python program to implement a Queue using array with ENQUEUE and DEQUEUE operations for data processing.




```python
queue = []

def enqueue(item):
    queue.append(item)
    print("Enqueued:", item)

def dequeue():
    if len(queue) == 0:
        print("Queue is empty")
    else:
        print("Dequeued:", queue.pop(0))

enqueue(1)
enqueue(2)
enqueue(3)
dequeue()
dequeue()
```

3. Singly Linked List

Write a Python program to implement a Singly Linked List with operations such as create, insert, delete, and traverse.




```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete(self):
        if self.head:
            self.head = self.head.next

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

ll = LinkedList()
ll.insert(10)
ll.insert(20)
ll.insert(30)
ll.traverse()
ll.delete()
ll.traverse()
```

4. Real-Life Application

Design and implement a mini project such as a Student Management System or Employee Management System using Python to perform:

Record creation

Data update

Calculations

Report generation

```python
students = []

def add_student(roll, name, marks):
    students.append([roll, name, marks])

def update_marks(roll, new_marks):
    for s in students:
        if s[0] == roll:
            s[2] = new_marks

def calculate_average():
    total = sum(s[2] for s in students)
    return total / len(students)

def report():
    for s in students:
        print("Roll:", s[0], "Name:", s[1], "Marks:", s[2])
    print("Average Marks:", calculate_average())

add_student(1, "Rahul", 80)
add_student(2, "Sneha", 90)
update_marks(1, 85)
report()
```

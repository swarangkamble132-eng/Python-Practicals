```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(10)
second = Node(20)
third = Node(30)

head.next = second
second.next = third

temp = head
while temp:
    print(temp.data)
    temp = temp.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

head = Node(20)
head.next = Node(30)

new_node = Node(10)
new_node.next = head
head = new_node

temp = head
while temp:
    print(temp.data)
    temp = temp.next
```

3. Count the number of digits in a given number

```python
number = 987654 # Example input
count = 0
temp_num = number

while temp_num != 0:
    temp_num //= 10  # Remove the last digit
    count += 1       # Increment the count

print(f"The number of digits in {number} is: {count}")
```

2. Find the sum of digits of a number

```python
number = 12345 
total_sum = 0
temp_num = number

while temp_num > 0:
    digit = temp_num % 10  
    total_sum += digit      
    temp_num //= 10         
print(f"The sum of digits of {number} is: {total_sum}")
```

3. Reverse a given number

```python
number = 12345 # Example input
reversed_num = 0
temp_num = number

while temp_num > 0:
    digit = temp_num % 10          # Get the last digit
    reversed_num = (reversed_num * 10) + digit # Append digit to reversed number
    temp_num //= 10                # Remove the last digit

print(f"The reverse of {number} is: {reversed_num}")
```

4. Check whether a number is a palindrome using a loop


```python
number = 121 # Example input
temp_num = number
reversed_num = 0

# Reverse the number using a while loop (as in the previous problem)
while temp_num > 0:
    digit = temp_num % 10
    reversed_num = (reversed_num * 10) + digit
    temp_num //= 10

# Check if original is equal to reversed
if number == reversed_num:
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")
```

1. Keep taking input from the user until the user enters `0`

```python
print("Enter numbers (enter 0 to stop):")

while True:
    try:
        user_input = int(input("Enter a number: "))
        if user_input == 0:
            print("0 entered. Stopping input.")
            break  # Exit the loop
        else:
            print(f"You entered: {user_input}")
    except ValueError:
        print("Invalid input. Please enter an integer.")
```

2. Create a menu-driven calculator (Addition, Subtraction, Multiplication, Division)

```python
def add(n1, n2): return n1 + n2
def subtract(n1, n2): return n1 - n2
def multiply(n1, n2): return n1 * n2
def divide(n1, n2):
    if n2 == 0:
        return "Cannot divide by zero"
    return n1 / n2

while True:
    print("\n--- Calculator Menu ---")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")
    choice = input("Enter choice (1-5): ")

    if choice in ['1', '2', '3', '4']:
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
            result = 0
            if choice == '1': result = add(num1, num2)
            elif choice == '2': result = subtract(num1, num2)
            elif choice == '3': result = multiply(num1, num2)
            elif choice == '4': result = divide(num1, num2)
            print(f"Result: {result}")
        except ValueError:
            print("Invalid number input.")
    elif choice == '5':
        print("Exiting calculator.")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

2. Find the factorial of a given number using a `for` loop


```python
num = 5  # Example number

if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    factorial = 1
    # Multiply factorial by i from 1 to num
    for i in range(1, num + 1):
        factorial = factorial * i
    print(f"The factorial of {num} is: {factorial}")
```

4. Check whether a given number is prime


```python
num = 13 # Example number to check

# Prime numbers are greater than 1
if num > 1:
    # Check for factors
    for i in range(2, int(num/2) + 1):
        if (num % i) == 0:
            print(f"{num} is not a prime number")
            break # If a factor is found, it's not prime, exit loop
    else: # This 'else' belongs to the for loop, runs if no break occurs
        print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

5. Count the number of vowels in a given string using a loop

```python
string = "Hello World" # Example string
vowel_count = 0
vowels = "aeiouAEIOU"

# Iterate over each character in the string
for char in string:
    if char in vowels:
        vowel_count += 1

print(f"The number of vowels in '{string}' is: {vowel_count}")
```

2	Calculate equivalent resistance of resistors connected in series and parallel

```python
#Series Connection
r1 = float(input("Enter R1: "))
r2 = float(input("Enter R2: "))
print("Series Resistance:", r1 + r2)

#Parallel Connection
print("Parallel Resistance:", (r1 * r2) / (r1 + r2))
```

3	Calculate voltage using Ohm’s Law

4	Check whether the entered frequency is audio or radio frequency

```python
f = float(input("Enter frequency: "))

if 20 <= f <= 20000:
    print("Audio Frequency")
else:
    print("Radio Frequency")
```

5	Display various radio frequency bands using if–elif ladder

```python
f = float(input("Enter frequency in MHz: "))

if f < 30:
    print("VLF Band")
elif f < 300:
    print("LF Band")
elif f < 3000:
    print("MF Band")
elif f < 30000:
    print("HF Band")
else:
    print("Above HF Band")
```

6	Display resistor color code using match-case

```python
num = int(input("Enter color code number (0-9): "))

match num:
    case 0: print("Black")
    case 1: print("Brown")
    case 2: print("Red")
    case 3: print("Orange")
    case 4: print("Yellow")
    case 5: print("Green")
    case 6: print("Blue")
    case 7: print("Violet")
    case 8: print("Grey")
    case 9: print("White")
    case _: print("Invalid code")
```

7	Demonstrate looping constructs

```python
arr = [10, 20, 30, 40]
key = int(input("Enter element: "))

found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Found at index", i)
        found = True
        break

if not found:
    print("Not Found")
```

```python
arr = [10, 20, 30, 40]
key = int(input("Enter element: "))

low = 0
high = len(arr) - 1

while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print("Found at index", mid)
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
else:
    print("Not Found")
```

```python
text = input("Enter string: ")
ch = input("Enter character to search: ")

for i in range(len(text)):
    if text[i] == ch:
        print("Found at position", i)
        break
else:
    print("Not Found")
```

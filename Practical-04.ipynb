```python
# Student Management System using List

student_marks = []
pass_mark = 40
distinction_mark = 75

while True:
    print("\n--- Student Management Menu ---")
    print("1. Add student mark")
    print("2. Display all marks")
    print("3. Total and Average marks")
    print("4. Highest and Lowest marks")
    print("5. Passed and Failed students")
    print("6. Distinction students")
    print("7. Remove duplicate marks")
    print("8. Apply grace marks")
    print("9. Search a mark")
    print("10. Sort marks and reverse")
    print("11. Remove a student record")
    print("12. Find position of a mark")
    print("0. Exit")

    choice = int(input("Enter your choice: "))

    # 1. Add mark
    if choice == 1:
        mark = int(input("Enter student mark: "))
        student_marks.append(mark)
        print("Mark added successfully")

    # 2. Display marks
    elif choice == 2:
        print("Student Marks:", student_marks)

    # 3. Total & Average
    elif choice == 3:
        if student_marks:
            total = sum(student_marks)
            average = total / len(student_marks)
            print("Total Marks:", total)
            print("Average Marks:", average)
        else:
            print("No records available")

    # 4. Highest & Lowest
    elif choice == 4:
        if student_marks:
            print("Highest Mark:", max(student_marks))
            print("Lowest Mark:", min(student_marks))
        else:
            print("No records available")

    # 5. Passed & Failed
    elif choice == 5:
        print("Passed Students:")
        for mark in student_marks:
            if mark >= pass_mark:
                print(mark)

        print("Failed Students:")
        for mark in student_marks:
            if mark < pass_mark:
                print(mark)

    # 6. Distinction
    elif choice == 6:
        print("Distinction Students:")
        for mark in student_marks:
            if mark >= distinction_mark:
                print(mark)

    # 7. Remove duplicates
    elif choice == 7:
        student_marks = list(set(student_marks))
        print("Duplicate marks removed")

    # 8. Grace marks
    elif choice == 8:
        for i in range(len(student_marks)):
            if student_marks[i] < pass_mark:
                student_marks[i] = pass_mark
        print("Grace marks applied")

    # 9. Search mark
    elif choice == 9:
        search_mark = int(input("Enter mark to search: "))
        if search_mark in student_marks:
            print("Mark found")
        else:
            print("Mark not found")

    # 10. Sort and reverse
    elif choice == 10:
        student_marks.sort()
        print("Ascending Order:", student_marks)
        print("Descending Order:", student_marks[::-1])

    # 11. Remove student record
    elif choice == 11:
        remove_mark = int(input("Enter mark to remove: "))
        if remove_mark in student_marks:
            student_marks.remove(remove_mark)
            print("Student record removed")
        else:
            print("Mark not found")

    # 12. Find position
    elif choice == 12:
        search_mark = int(input("Enter mark: "))
        if search_mark in student_marks:
            print("Position:", student_marks.index(search_mark))
        else:
            print("Mark not found")

    # Exit
    elif choice == 0:
        print("Exiting Program")
        break

    else:
        print("Invalid choice")
```

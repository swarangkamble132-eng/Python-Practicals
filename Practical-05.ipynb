```python
books = tuple(map(int, input("Enter book values: ").split()))

print("Books:", books)
print("Total books:", len(books))
print("Average value:", sum(books) / len(books))
print("Highest value:", max(books))
print("Lowest value:", min(books))

limit = int(input("Enter premium limit: "))
print("Premium books:", [b for b in books if b >= limit])

search = int(input("Enter value to search: "))
print("Found" if search in books else "Not Found")
print("Count:", books.count(search))

books = tuple(set(books))
print("After removing duplicates:", books)

print("Reversed tuple:", books[::-1])

value = int(input("Enter value to find index: "))
print("Index:", books.index(value) if value in books else "Value not present")
```

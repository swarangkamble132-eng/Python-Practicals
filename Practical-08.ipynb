```python
def half_adder(a, b):
    sum = a ^ b
    carry = a & b
    print("Half Adder -> Sum:", sum, "Carry:", carry)

def full_adder(a, b, c):
    sum = a ^ b ^ c
    carry = (a & b) | (b & c) | (a & c)
    print("Full Adder -> Sum:", sum, "Carry:", carry)

half_adder(1, 1)
full_adder(1, 1, 1)
```

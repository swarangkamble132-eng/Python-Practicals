```python
cart = set(map(int, input("Enter product IDs: ").split()))

print("Cart items:", cart)
print("Total products:", len(cart))
print("Highest ID:", max(cart))
print("Lowest ID:", min(cart))

search = int(input("Enter product ID to search: "))
print("Found" if search in cart else "Not Found")

remove_id = int(input("Enter product ID to remove: "))
cart.discard(remove_id)
print("After removal:", cart)

limit = int(input("Enter premium limit: "))
print("Premium products:", {p for p in cart if p >= limit})

cart2 = set(map(int, input("Enter second cart IDs: ").split()))
print("Union:", cart | cart2)
print("Intersection:", cart & cart2)

print("Sorted cart:", sorted(cart))
```

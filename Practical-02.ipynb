Program to display RF bands using if–elif

```python
f = float(input("Enter frequency in Hz: "))

if 30000 <= f <= 300000:
    print("VLF Band")
elif 300000 < f <= 3000000:
    print("LF Band")
elif 3000000 < f <= 30000000:
    print("MF Band")
elif 30000000 < f <= 300000000:
    print("HF Band")
elif 300000000 < f <= 3000000000:
    print("VHF Band")
else:
    print("UHF / Higher Band")
```

Write a Python program that checks if a resistor value is within a tolerance range (e.g., 5% tolerance for a 100Ω resistor) and prints whether it is within tolerance or not.

```python
value = float(input("Enter measured resistance: "))

min_val = 100 - (5/100)*100
max_val = 100 + (5/100)*100

if min_val <= value <= max_val:
    print("Resistor is within tolerance")
else:
    print("Resistor is out of tolerance")
```

```python
Q1: Largest of two numbers
IF a > b THEN
    PRINT "A is largest"
ELSE
    IF b > a THEN
        PRINT "B is largest"
    ELSE
        PRINT "Both equal"
    ENDIF
ENDIF
```

```python
a = float(input("Enter number a: "))
b = float(input("Enter number b: "))
if a > b:
    print("A is largest")
else:
    if b > a:
        print("B is largest")
    else:
        print("Both equal")
```

```python
Q2: Result with distinction
IF marks >= 40 THEN
    IF marks >= 75 THEN
        PRINT "Distinction"
    ELSE
        PRINT "Pass"
    ENDIF
ELSE
    PRINT "Fail"
ENDIF
```

```python
marks = float(input("Enter marks: "))
if marks >= 40:
    if marks >= 75:
        print("Distinction")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
Q3: Login and role check
IF login = TRUE THEN
    IF role = "Admin" THEN
        PRINT "Admin Access"
    ELSE
        PRINT "User Access"
    ENDIF
ELSE
    PRINT "Login Failed"
ENDIF
```

```python
login = input("Are you logged in? (True/False): ").lower() == 'true'
role = input("Enter your role: ")

if login == True:
    if role == "Admin":
        print("Admin Access")
    else:
        print("User Access")
else:
    print("Login Failed")
```

```python
Q4: Weather advice
IF raining = TRUE THEN
    IF umbrella = TRUE THEN
        PRINT "Go Outside"
    ELSE
        PRINT "Stay Inside"
    ENDIF
ELSE
    PRINT "Enjoy the Day"
ENDIF
```

```python
is_raining = input("Is it raining? (True/False): ").lower() == 'true'
has_umbrella = input("Do you have an umbrella? (True/False): ").lower() == 'true'

if is_raining and has_umbrella:
    print("Go Outside")
elif is_raining and not has_umbrella:
    print("Stay Inside")
else:
    print("Enjoy the Day")
```

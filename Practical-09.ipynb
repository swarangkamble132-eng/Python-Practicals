```python
class Student:
    def get_data(self):
        self.roll = 1
        self.name = "Amit"
        self.course = "Computer Engineering"
        self.percentage = 85.40
        self.college = "Vidyalankar Polytechnic"

    def display(self):
        print("Roll No:", self.roll)
        print("Name:", self.name)
        print("Course:", self.course)
        print("Percentage:", self.percentage)
        print("College:", self.college)

s = Student()
s.get_data()
s.display()
```

```python
class Student:
    def __init__(self, roll, name, course, percentage, city):
        self.roll = roll
        self.name = name
        self.course = course
        self.percentage = percentage
        self.city = city

    def display(self):
        print(self.roll, self.name, self.course, self.percentage, self.city)

s = Student(2, "Neha", "IT Engineering", 88.90, "Mumbai")
s.display()
```

```python
class Student:
    def __init__(self, roll, name, course, percentage):
        self.roll = roll
        self.name = name
        self.course = course
        self.percentage = percentage

    def display(self):
        print(self.roll, self.name, self.course, self.percentage)

s1 = Student(1, "Rahul", "Computer", 82)
s2 = Student(2, "Sneha", "IT", 90)
s3 = Student(3, "Aakash", "AI", 78)

s1.display()
s2.display()
s3.display()
```
